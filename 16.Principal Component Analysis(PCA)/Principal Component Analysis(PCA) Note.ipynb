<center><h1 style="color:green">Principal Component Analysis(PCA)</center>

# Why PCA?
High-dimensional datasets often suffer from the **curse of dimensionality**, leading to:
1. **Overfitting**: Models fail to generalize to new data.
2. **Increased computational complexity**: Processing becomes expensive.
3. **Reduced accuracy**: Algorithms struggle to perform effectively.

To address these, we use **dimensionality reduction** techniques like PCA, which:
- Reduce the number of features while retaining maximum variance.
- Simplify data for visualization, modeling, and interpretation.

---

# What is PCA?
PCA, introduced by Karl Pearson in 1901, is a statistical method that:
- **Transforms correlated variables** into a set of uncorrelated variables (principal components).
- **Identifies the axes of maximum variance** in the data.
- Reduces the dimensionality of datasets while preserving essential patterns and relationships.

---

# Key Features of PCA
1. **Unsupervised Learning Algorithm**: Does not rely on target variables.
2. **Applications**:
   - **Data Visualization**: Reduces data to 2 or 3 dimensions for easier plotting.
   - **Feature Selection**: Identifies significant variables.
   - **Data Compression**: Reduces dataset size without significant information loss.
3. **Assumption**: Variance carries most of the information in the data.

---

# Steps in PCA

### **1. Standardization**
Standardize the dataset to have a mean of 0 and a standard deviation of 1:

$$ Z = \frac{X - \mu}{\sigma} $$

Where:
- $ \mu $: Mean of features.
- $ \sigma $: Standard deviation of features.

---

### **2. Compute Covariance Matrix**
The covariance matrix captures the joint variability between features:

$$ \text{cov}(x_1, x_2) = \frac{\sum_{i=1}^n (x_{1i} - \bar{x}_1)(x_{2i} - \bar{x}_2)}{n-1} $$

Covariance indicates:
- **Positive**: Variables increase together.
- **Negative**: One variable increases while the other decreases.
- **Zero**: No relationship.

---

### **3. Eigenvalues and Eigenvectors**
Solve for eigenvalues ($ \lambda $) and eigenvectors ($ X $) of the covariance matrix:

$$ AX = \lambda X $$

Where:
- $ A $: Covariance matrix.
- $ \lambda $: Eigenvalues (variance captured by each principal component).
- $ X $: Eigenvectors (directions of principal components).

Eigenvalues and eigenvectors are computed as:

$$ |A - \lambda I| = 0 $$

Where $ I $ is the identity matrix of the same dimension as $ A $.

---

# Principal Components
1. **First Principal Component**: Captures the maximum variance in the dataset.
2. **Subsequent Components**: Capture remaining variance, orthogonal to earlier components.

---

# Benefits of PCA
1. **Simplifies Complex Data**: Helps in understanding high-dimensional datasets.
2. **Improves Model Performance**: Reduces risk of overfitting.
3. **Reduces Computational Cost**: Less data to process.

PCA is a powerful technique for dimensionality reduction and exploratory data analysis, making it indispensable in machine learning and data science.
