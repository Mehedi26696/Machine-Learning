```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop('Survived',axis='columns')
```

```python
inputs
```

```python
target
```

```python
dummies = pd.get_dummies(inputs.Sex).astype(int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

<b>I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

<b>Calculate the score using cross validation

```python
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
